## Introduction 

The goal of this project is to scrape the web for data to compare the sentiment of the websites towards the iPhone and the Samsung Galaxy to recommend the client one of the two phones where they should deploy the app they're working on.

To do that, the web was scrapped from Common Crawl using AWS, looking for positive and negative words around some keywords referring to 5 different mobile phones.

To know whether the count of positive and negative words near the different phones' attributes result in an overall positive or negative sentiment toward a specific device (in this case Iphone and Samsung Galaxy, the sentiment of which will be measured on a scale from 0 to 5), a dataset where the sentiment has been manually labeled is provided for the Iphone and the Samsung Galaxy, both datasets also containing the same attributes, namely the same word counts.

The training data (the two datasets manually labeled) will be preprocessed, featured selected, featured engineered and modelled to provide an accurate prediction for the data scrapped from the web. This prediction will allow to know whether the general sentiment over one phone prevails over the other.

#### Dealing with class imbalance

##### Downsample the size of the highest class to the size of the second higest class

To deal with the class imbalance, one of the target variables will be chosen (_iphonesentiment_ in this case) and the highest rating ("5") will be downsampled to the size of the second highest ("0"), since there is such a huge difference between the highest and the second highest classes, but not a big difference among the other classes.

```python
# Undersample the data to have the highest rating ("5") match the size of the second highest rating ("0")
iphone_sentiment = df.groupby('iphonesentiment').size()
dict_resample = {0:iphone_sentiment.iloc[0],
                 1:iphone_sentiment.iloc[1],
                 2:iphone_sentiment.iloc[2],
                 3:iphone_sentiment.iloc[3],
                 4:iphone_sentiment.iloc[4],
                 5:iphone_sentiment.sort_values(ascending = False).iloc[1]}
rus_highest = RandomUnderSampler(random_state=0, sampling_strategy = dict_resample)
X_resampled, y_resampled = rus_highest.fit_resample(df.iloc[:,:-2], df['iphonesentiment'])
```

After resampling the data for _iphonesentiment_, the indices that are saved for the undersampled dataframe are used to create a new dataframe from the original one, containig both _iphonesentiment_ and _galaxysentiment_. Then the distribution of both of them will be plotted to see if the undersampling also worked for _galaxysentiment_.

```python
# Create a new dataframe with the samples that have been kept after downsampling
df_resampled_highest = df.iloc[rus_highest.sample_indices_,:]


# Distribution of the iphone sentiment after resampling
df_resampled_highest.groupby('iphonesentiment').size().plot(kind='bar')
plt.xlabel('Iphone sentiment')
plt.ylabel('Frequency')
plt.title('Number of websites for each iphone sentiment value')
plt.show()


# Distribution of the galaxy sentiment after resampling
df_resampled_highest.groupby('galaxysentiment').size().plot(kind='bar')
plt.xlabel('Galaxy sentiment')
plt.ylabel('Frequency')
plt.title('Number of websites for each galaxy sentiment value')
plt.show()
```

The distributions for both datasets seems decent. However, another downsampling technique will be applied, where all the values of the sentiment will be downsampled to the lowest category. After, in the modelling stage, both undersampling techniques will be compared, as well as comparing to the data with no class imbalance treatment.

##### Downsampling every class size to the size of the lowest class size

```python
# Undersample the data to have the all the classes with the same size (the size of the lowest class count)
rus_all = RandomUnderSampler(random_state=0, sampling_strategy = 'all')
X_resampled, y_resampled = rus_all.fit_resample(df.iloc[:,:-2], df['iphonesentiment'])


# Create a new dataframe with the samples that have been kept after downsampling
df_resampled_all = df.iloc[rus_all.sample_indices_,:]


# Distribution of the iphone sentiment after resampling
df_resampled_all.groupby('iphonesentiment').size().plot(kind='bar')
plt.xlabel('Iphone sentiment')
plt.ylabel('Frequency')
plt.title('Number of websites for each iphone sentiment value')
plt.show()


# Distribution of the galaxy sentiment after resampling
df_resampled_all.groupby('galaxysentiment').size().plot(kind='bar')
plt.xlabel('Galaxy sentiment')
plt.ylabel('Frequency')
plt.title('Number of websites for each galaxy sentiment value')
plt.show()
```

##### Oversampling every class size to the size of the highest class size

One last resampling strategy is to oversample, where the minority classes will randomly be repeated (using bootstrap) so that all classes are brought to the level of the majority class.

```python
# Oversample the data to have the all the classes with the same size (the size of the highest class count)
ros_all = RandomOverSampler(random_state=0, sampling_strategy = 'all')
X_resampled, y_resampled = ros_all.fit_resample(df.iloc[:,:-2], df['iphonesentiment'])


# Create a new dataframe with the samples that have been kept after downsampling
df_resampled_all = df.iloc[ros_all.sample_indices_,:]


# Distribution of the iphone sentiment after resampling
df_resampled_all.groupby('iphonesentiment').size().plot(kind='bar')
plt.xlabel('Iphone sentiment')
plt.ylabel('Frequency')
plt.title('Number of websites for each iphone sentiment value')
plt.show()


# Distribution of the galaxy sentiment after resampling
df_resampled_all.groupby('galaxysentiment').size().plot(kind='bar')
plt.xlabel('Galaxy sentiment')
plt.ylabel('Frequency')
plt.title('Number of websites for each galaxy sentiment value')
plt.show()
```

### Correlation

The first approach to feature selection is correlation. In this case, specifically, the independent variables will be correlated to each other and high levels of correlation (collinearity) will be removed based on common sense.  
Two alternatives will be taken here, the first one will be to remove levels of collinearity higher than 0.9, and the other one to remove levels of collinearity higher than 0.95.

```python
# Compute correlation of iphone and galaxy datasets (this will exclude the target variable since it's categorical)
corr = df_predictors.corr()


# Turn the correlation matrix into a pairwise dataframe of correlations
corr_upper = pd.DataFrame(np.triu(corr))
corr_upper.columns = corr.columns
corr_upper.index = corr.index
corr_upper = corr_upper.unstack()


# Erase the autocorrelations from the pairwise dataframe of correlations
for index in corr_upper.index :
    if(index[0] == index[1]) :
        corr_upper = corr_upper.drop(index, axis = 0)


# Select only correlations that have high values
corr_upper_highcorr095 = pd.DataFrame(corr_upper[np.logical_or((corr_upper > 0.95).values, (corr_upper < -0.95).values)])
corr_upper_highcorr095.columns = ['correlation']
print('High correlated pariwise values (> 0.95) - multiindex form:')
display(corr_upper_highcorr095)


# Convert from multiindex to a dataframe of 3 columns, 2 of the correlated attributes and 1 with the correlation value
#corr_upper_highcorr095 = corr_upper_highcorr095.reset_index()
#corr_upper_highcorr095.columns = ['attribute_1', 'attribute_2', 'correlation']
#print('High correlated pariwise values (> 0.95) - column form:')
#display(corr_upper_highcorr095)


# Select only correlations that have high values
corr_upper_highcorr090 = pd.DataFrame(corr_upper[np.logical_or((corr_upper > 0.90).values, (corr_upper < -0.90).values)])
corr_upper_highcorr090.columns = ['correlation']
print('High correlated pariwise values (> 0.90) - multiindex form:')
display(corr_upper_highcorr090)


# Convert from multiindex to a dataframe of 3 columns, 2 of the correlated attributes and 1 with the correlation value
#corr_upper_highcorr090 = corr_upper_highcorr090.reset_index()
#corr_upper_highcorr090.columns = ['attribute_1', 'attribute_2', 'correlation']
#print('High correlated pariwise values (> 0.90) - column form:')
#display(corr_upper_highcorr090)
```

#### Remove highly collinear variables and save the datasets

Based on the results in the tables above, a few attributes will be removed from the dataset (those that seem less relevant in the pairwise high collinear values).  
In this case, the removed attributes will be the following:

### Variance

The second approach to feature selection is variance. In this case, specifically, the variance of the independent variables will be calculated and a scree plot will be obtained, based on which, a number of attributes will be removed.

```python
# Calculate the variance of the predictors and set it into a percentage of the total variance, sorted from highest to lowest
df_variance = df_predictors.var()
df_variance_percentage = df_variance / df_variance.sum() * 100
df_variance_percentage.sort_values(ascending = False, inplace=True)


# Scree plot of the total percentage of explained variance
plt.plot(np.arange(1,(len(df_variance_percentage)+1)), np.cumsum(df_variance_percentage),
         linestyle='--', marker='o', color='b')

plt.title("Cumulative explained variance of the dataset attributes")
plt.xlabel('Number of attributes')
plt.ylabel('Cumulative explained variance')
plt.show()


# Show the cumulative explained variance percentage
print('Cumulative variance percentage:')
cumsum_percentage = np.cumsum(df_variance_percentage)
for i, (name, value) in enumerate(cumsum_percentage.items()) :
    print('Cumulative variance of attribute #%d with name %s = %.2f' % (i+1, name, value) + '%')
```

#### Remove low variance variables and save the datasets

From the scree plot and the cumulative variance it is seen that setting the variance of the attribute _iphonecamneg_ as a limit and removing attributes with an equal or lower variance 90% of the explained variance will be kept, which will keep 13 attributes. And by setting the variance of the attribute _samsungdisunc_ as a limit and removing attributes with an equal or lower variance 98% of the explained variance will be kept, which will keep 30 attributes.


```python
# Keep attributes that explain 90% of the variance
variance090 = df_variance['iphonecamneg']
variance_selector = VarianceThreshold(threshold = variance090)
variance_selector.fit(df_predictors)
# Get the columns that have variance higher than the limit set and add the 2 target variables to create a new dataset
df_variance090 = pd.concat([df[df.columns[variance_selector.get_support(indices=True)]], df[['iphonesentiment','galaxysentiment']]], axis = 1)

# Keep attributes that explain 98% of the variance
variance098 = df_variance['samsungdisunc']
variance_selector = VarianceThreshold(threshold = variance098)
variance_selector.fit(df_predictors)
df_variance098 = pd.concat([df[df.columns[variance_selector.get_support(indices=True)]], df[['iphonesentiment','galaxysentiment']]], axis = 1)
```

```python
# Fit the PCA data
pca = PCA(n_components=None)
pca.fit(df_predictors)


# Scree plot of the PCA
cumsum_percentage_pca = np.cumsum(pca.explained_variance_ratio_) * 100
plt.plot(np.arange(1,(len(df_predictors.columns)+1)), cumsum_percentage_pca,
         linestyle='--', marker='o', color='b')

plt.title("Cumulative explained variance of the PCA")
plt.xlabel('number of principal components')
plt.ylabel('Cumulative explained variance')
plt.show()


# Show the cumulative explained variance percentage of the PCA
print('Cumulative variance percentage of the PCA:')
for i, k in enumerate(cumsum_percentage_pca) :
    print('Cumulative variance of element #%d = %.2f' % (i+1, k) + '%')
```

```python
# Fit and transform the 6 attributes that explain most of the variance of the PCA (90% of the variance)
number_of_components = 6
pca = PCA(n_components = number_of_components)  #Number of components to keep. Default -> all components kept
pca.fit(df_predictors)
pc_data = pca.transform(df_predictors)


# Eliminate the predictor columns and substitute them by the 6 components of the PCA
df_pca090 = df.copy()
df_pca090 = df_pca090.drop(df[df_predictors.columns], axis = 1)
predictors_pca = ['PCA0'+ str(x) for x in range(0,number_of_components)] ###
df_pca090[predictors_pca] = pd.DataFrame(pc_data)
# Reorder so that the 2 target variables are at the end instead of at the beginning
df_pca090 = df_pca090[df_pca090.columns[2:].append(df_pca090.columns[0:2])]


# Fit and transform the 16 attributes that explain most of the variance of the PCA (98% of the variance)
number_of_components = 16
pca = PCA(n_components = number_of_components)  #Number of components to keep. Default -> all components kept
pca.fit(df_predictors)
pc_data = pca.transform(df_predictors)


# Eliminate the predictor columns and substitute them by the 6 components of the PCA
df_pca098 = df.copy()
df_pca098 = df_pca098.drop(df[df_predictors.columns], axis = 1)
predictors_pca = ['PCA0'+ str(x) for x in range(0,number_of_components)] ###
df_pca098[predictors_pca] = pd.DataFrame(pc_data)
# Reorder so that the 2 target variables are at the end instead of at the beginning
df_pca098 = df_pca098[df_pca098.columns[2:].append(df_pca098.columns[0:2])]
```

### Engineering the dependant variable

The second approach to feature engineering is engineering the dependant variable. In this case, specifically, the dependent variables (iphonesentiment and galaxysentiment) will be reduced from a scale of 0 to 6 (6 different categories) to a scale of 0 to 4 (4 different categories) by combining the two lowest values (0 and 1), as well as the two highest values (4 and 5), in the same category.  

Since this approach doesn't overlap with the other approaches, it can be added to them. To do this, all the datasets created until this point will be taken and two new columns will be added containing the new category scales.  

This means that there will be 4 target columns (2 iphonesentiment and 2 galaxysentiment, each of them with different scales of sentiment) and each of them will be tested in the modelling part.

```python
# Create a dictionary with the target values as keys as and a dictionary as value. This last dicionary will contain 
# the values to replace as keys and the values to be replaced for as values
dict_replace = {'iphonesentiment':{0:1, 5:4}, 'galaxysentiment':{0:1, 5:4}}
# Replace the values created with the dicionary and store the result
# (the two columns with the new target variables) in a new dataset
target_variables_less_categories = df[['iphonesentiment','galaxysentiment']].replace(dict_replace)
```

```python
# Loop through all the saved datasets in the df_list dictionary and add to each of them two columns
# (the columns added will be the two target variables stored in df_less_categories,
# which are the target variables recoded to have less categories)
for key, value in df_dict.items() :
    value.loc[:,'iphonesentiment_2'] = target_variables_less_categories['iphonesentiment']
    value.loc[:,'galaxysentiment_2'] = target_variables_less_categories['galaxysentiment']
```

## Model development and evaluation

Four different modelling systems will be implemented.  
* The first one with the target variable being _iphonesentiment_ on a scale from 0 to 6.
* The second one with the target variable being _iphonesentiment_ on a scale from 1 to 5.
* The third one with the target variable being _galaxysentiment_ on a scale from 0 to 6.
* The first one with the target variable being _galaxysentiment_ on a scale from 1 to 5.  

Each of the modelling systems will be tested for each of the dataframes saved during the feature selection and the feature engineering stages and each of them will be fit with different models. Then the best accuracy will be kept for each of the four modelling systems.  

Before that, however, the models to be used will be defined in a list.

```python
# Define the models to be used
models = []
models.append(('LR', LogisticRegression(solver='liblinear',
                                        multi_class='ovr')))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('DT', DecisionTreeClassifier()))
models.append(('RF', RandomForestClassifier()))


# Get the names of the models
names_of_models = []
for name, model in models :
    names_of_models.append(name)
```

### Iphone sentiment predictions (scale from 0 to 6)

#### Separate train and test

```python
# Separate predictors and targets
X_iphonesentiment_train = {}
y_iphonesentiment_train = {}
X_iphonesentiment_validation = {}
y_iphonesentiment_validation = {}
for name, df in df_dict.items() :
    X_train, X_validation, Y_train, Y_validation = train_test_split(df[df.columns[0:-4]], df[df.columns[-4]],
                                                                test_size = 0.20,
                                                                random_state = 1,
                                                                shuffle = True)
    X_iphonesentiment_train.update({name : X_train})
    y_iphonesentiment_train.update({name : Y_train})
    X_iphonesentiment_validation.update({name : X_validation})
    y_iphonesentiment_validation.update({name : Y_validation})    
```

#### Evaluate the models

```python
# Evaluate each model in each dataframe and store the accuracy of the CV in the results dataframe
results_iphone_large_scale_cv = pd.DataFrame(columns = names_of_models, index = list(df_dict.keys()))
results_iphone_large_scale_test = pd.DataFrame(columns = names_of_models, index = list(df_dict.keys()))
for df_name in df_dict.keys() :
    model_results = {}
    for model_name, model in models :
        kfold = StratifiedKFold(n_splits=10, random_state=1)
        cv_results = cross_val_score(model, X_iphonesentiment_train[df_name], y_iphonesentiment_train[df_name], 
                                     cv=kfold, scoring='accuracy')
        results_iphone_large_scale_cv.loc[df_name, model_name] = cv_results.mean()
        model.fit(X_iphonesentiment_train[df_name], y_iphonesentiment_train[df_name])
        predictions = model.predict(X_iphonesentiment_validation[df_name])
        results_iphone_large_scale_test.loc[df_name, model_name] = metrics.accuracy_score(y_iphonesentiment_validation[df_name], predictions)
```

```python
# Show the results for cross-validation and test
print('Cross-Validation accuracy:')
display(results_iphone_large_scale_cv)
print('Test accuracy:')
display(results_iphone_large_scale_test)
```

Since the results of the Cross-Validation and the results of the Validation set are very similar, only one of them will be taken into account for the final comparison (in this case the Cross-Validation accuracy).

#### Choose the best model

```python
# Find the max value of accuracy and drop all other values to know what dataframe and model it corresponds to
best_result_iphone_large_scale = results_iphone_large_scale_cv.where(results_iphone_large_scale_cv == results_iphone_large_scale_cv.values.max()).dropna(how='all').dropna(axis=1)
display(best_result_iphone_large_scale)
```

For the Iphone sentiment prediction using a scale from 0 to 6, the best dataframe is the one where the collinear attributes with a correlation higher than 90% are removed, using a Random Forest model. The accuracy for this is 76%.

```python
# Evaluate each model in each dataframe and store the accuracy of the CV in the results dataframe
results_galaxy_small_scale_cv = pd.DataFrame(columns = names_of_models, index = list(df_dict.keys()))
results_galaxy_small_scale_test = pd.DataFrame(columns = names_of_models, index = list(df_dict.keys()))
for df_name in df_dict.keys() :
    model_results = {}
    for model_name, model in models :
        kfold = StratifiedKFold(n_splits=10, random_state=1)
        cv_results = cross_val_score(model, X_galaxysentiment2_train[df_name], y_galaxysentiment2_train[df_name], 
                                     cv=kfold, scoring='accuracy')
        results_galaxy_small_scale_cv.loc[df_name, model_name] = cv_results.mean()
        model.fit(X_galaxysentiment2_train[df_name], y_galaxysentiment2_train[df_name])
        predictions = model.predict(X_galaxysentiment2_validation[df_name])
        results_galaxy_small_scale_test.loc[df_name, model_name] = metrics.accuracy_score(y_galaxysentiment2_validation[df_name], predictions)
```

```python
# Show the results for cross-validation and test
print('Cross-Validation accuracy:')
display(results_galaxy_small_scale_cv)
print('Test accuracy:')
display(results_galaxy_small_scale_test)
```

Since the results of the Cross-Validation and the results of the Validation set are very similar, only one of them will be taken into account for the final comparison (in this case the Cross-Validation accuracy).

#### Choose the best model

```python
# Find the max value of accuracy and drop all other values to know what dataframe and model it corresponds to
best_result_galaxy_small_scale = results_galaxy_small_scale_cv.where(results_galaxy_small_scale_cv == results_galaxy_small_scale_cv.values.max()).dropna(how='all').dropna(axis=1)
display(best_result_galaxy_small_scale)
```
